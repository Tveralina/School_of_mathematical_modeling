```python
import numpy as np
from decimal import *
getcontext().prec = 2000
```

```python
# корни многочлена
roots = np.linspace(1, 2021, num=2021, endpoint=True, dtype='int')
roots
```

```python
# промежуточные значения по x (без целых чисел)
x_arr = np.linspace(1.0000001, 2021, num=1000000, endpoint=False)
y_arr = np.zeros_like(x_arr)
```

```python
# получение суммы значение (1/(x-1)), где i = (1, .., 2021)
for i in range(y_arr.shape[0]):
    y_arr[i] = (1 / (x_arr[i] - roots)).sum()

y_arr
```

```python
# получение индексов значений по X в которых производная близка к 0
x_zero = np.isclose(y_arr, np.zeros_like(y_arr), atol=0.00001).nonzero()
x_zero
```

```python
# вывод значений по X в которых производная равна 0
print(f'x1: {x_arr[x_zero[0][0]]}, x2: {x_arr[x_zero[0][1]]}')
```

```python
# аккумулируем значения каждого множителя (x - i) / 2021, где i = от 1 до 2021
x1 = Decimal(x_arr[x_zero[0][0]])
x2 = Decimal(x_arr[x_zero[0][1]])
dec_number1 = Decimal(1)
dec_number2 = Decimal(1)
for i in range(1, 2022):
    dec_number1 *= (x1 - i) / 2021
    dec_number2 *= (x2 - i) / 2021

print(dec_number1)
print(dec_number2)
```

```python
# вывод значения
dec_number1.as_tuple()
```

```python
# переводим значения digit в целое число
int_number1 = int(''.join([str(x) for x in list(dec_number1.as_tuple()[1])]))
int_number2 = int(''.join([str(x) for x in list(dec_number2.as_tuple()[1])]))
```

```python
# умножаем int_number на 2021^2021 (потому что до этого мы делили на 2021)
big_number1 = int_number1 * (2021**2021)
big_number2 = int_number2 * (2021**2021)
```

```python
# переводим big_number в строку, откидываем последние числа (столько, сколько указано в exponent для числа dec_number)
extremum1 = int(str(big_number1)[:-(abs(dec_number1.as_tuple()[2]) + len(str(int_number1)))])
extremum2 = int(str(big_number2)[:-(abs(dec_number2.as_tuple()[2]) + len(str(int_number2)))])
```

```python
if extremum1 > extremum2:
    print(f'Минимальное значение многочлена в точке x = {x_arr[x_zero[0][1]]} равно:\ny = {extremum2}\nКол-во знаков y: {len(str(extremum2))}')
    print(f'Максимальное значение многочлена в точке x = {x_arr[x_zero[0][0]]} равно:\ny = {extremum1}\nКол-во знаков y: {len(str(extremum1))}')
    print(f'Амплитуда значений Ymax-Ymin = {extremum1-extremum2}\nКол-во знаков Ymax-Ymin: {len(str(extremum1-extremum2))}')
else:
    print(f'Минимальное значение многочлена в точке x = {x_arr[x_zero[0][0]]} равно:\ny = {extremum1}\nКол-во знаков y: {len(str(extremum1))}')
    print(f'Максимальное значение многочлена в точке x = {x_arr[x_zero[0][1]]} равно:\ny = {extremum2}\nКол-во знаков y: {len(str(extremum2))}')
    print(f'Амплитуда значений Ymax-Ymin = {extremum2-extremum1}\nКол-во знаков Ymax-Ymin: {len(str(extremum2-extremum1))}')
```
